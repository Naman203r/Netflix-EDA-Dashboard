# **Data Cleaning**

```python
# Remove Leading and Trailing Spaces in Categorical Columns
netflix_data['show_id']=netflix_data['show_id'].str.strip()
netflix_data['type']=netflix_data['type'].str.strip()
netflix_data['title']=netflix_data['title'].str.strip()
netflix_data['director']=netflix_data['director'].str.strip()
netflix_data['cast']=netflix_data['cast'].str.strip()
netflix_data['country']=netflix_data['country'].str.strip()

netflix_data['date_added']=netflix_data['date_added'].str.strip()
netflix_data['rating']=netflix_data['rating'].str.strip()
netflix_data['duration']=netflix_data['duration'].str.strip()
netflix_data['listed_in']=netflix_data['listed_in'].str.strip()
netflix_data['description']=netflix_data['description'].str.strip()
```

```python
# Rename Columns
netflix_data=netflix_data.rename(columns={'show_id':'content_id','cast':'actor','listed_in':'genre'})
```

```python
# Handle Missing Values
netflix_data['director']=netflix_data['director'].fillna('Unknown Director')
netflix_data['actor']=netflix_data['actor'].fillna('Unknown Actor')
netflix_data['country']=netflix_data['country'].fillna("Unknown Country")
netflix_data['date_added']=netflix_data['date_added'].fillna("Unknown Date")
netflix_data['rating']=netflix_data['rating'].fillna("Unknown Rating")
netflix_data['duration']=netflix_data['duration'].fillna("Unknown Duration")
```

```python
# Replace character 's' with 'c' in 'content_id' column
netflix_data['content_id']=netflix_data['content_id'].str.replace('^s','C',regex=True)
```

```python
# Optimize Column Data Types to Save Memory
netflix_data['type']=netflix_data['type'].astype('category')
netflix_data['country']=netflix_data['country'].astype('category')
netflix_data['release_year']=netflix_data['release_year'].astype(np.int16)
netflix_data['rating']=netflix_data['rating'].astype('category')
netflix_data['duration']=netflix_data['duration'].astype('category')
netflix_data['genre']=netflix_data['genre'].astype('category')
```

```python
# Remove 'description' column as it is unnecessary for EDA/Dashboard
netflix_data = netflix_data.drop(columns=['description'])
```

# **Data Transformation**

```python
# Columns with Multiple Values
multi_cols=['director', 'actor', 'country', 'genre']

# Split Strings into Lists
for col in multi_cols:
    netflix_data[col]=netflix_data[col].str.split(', ')

# Explode Each Column one by one
for col in multi_cols:
    netflix_data=netflix_data.explode(col)

# Strip leading/trailing Spaces
for col in multi_cols:
    netflix_data[col]=netflix_data[col].str.strip()

# Reset index
netflix_data=netflix_data.reset_index(drop=True)
```

# Feature Engineering on duration column

```python
# For movie
def check_minute(value):
  if 'min' in value:
    return int(value.replace(' min',''))
  else:
    return np.nan

# For Session
def check_Season(value):
  if 'Season' in value:
    return int(value.split(' ')[0])
  else:
    return np.nan

clean_data['minute']=clean_data['duration'].apply(check_minute)
clean_data['Season']=clean_data['duration'].apply(check_Season)
```

```python
clean_data.sample(5)
```

# Univariate Analysis on minute column
# Distribution Insight:
- Movie durations follow a slightly right-skewed distribution.
- Most movies have durations between 87 and 114 minutes.
- A few movies are very long (200+ minutes), which stretch the right tail of the distribution.
- Extremely long movies (> 240 minutes) are rare.

# Outlier Detection Insight:

- Using IQR, the normal range of movie durations is 87114 minutes.
- Movies with durations above 150 minutes are considered statistical outliers.

```python
clean_data['minute'].describe()
```

```python
clean_data['minute'].plot(kind='box')
```

```python
clean_data['minute'].plot(kind='kde')
```

```python
# Detect outliers
Q1_minute=clean_data['minute'].quantile(0.25)
Q3_minute=clean_data['minute'].quantile(0.75)
IQR_minute=Q3_minute-Q1_minute
lower_boundary_minute=Q1_minute-1.3*IQR_minute
upper_boundary_minute=Q3_minute+1.3*IQR_minute
print(lower_boundary_minute)
print(upper_boundary_minute)
clean_data['minute'][(clean_data['minute']<lower_boundary_minute) | (clean_data['minute']>upper_boundary_minute)]
```
